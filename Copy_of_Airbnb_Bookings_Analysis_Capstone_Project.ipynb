**2. What can we learn from predictions? (ex: locations, prices, reviews, etc)?**

```python
areas_reviews=new_df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_reviews
```

```python
area = areas_reviews['neighbourhood_group']
review = areas_reviews['number_of_reviews']
fig = plt.figure(figsize=(10,5))

#creating the bar plot
plt.bar(area, review, color='yellow',width=0.4)

plt.xlabel('area')
plt.ylabel('review')
plt.title('Area vs Number of reviews')
plt.show()
```

```python
price_area = new_df.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(5)
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

fig = plt.figure(figsize = (10,5))

# creat bar plot
plt.scatter(area, price)

plt.xlabel('Price')
plt.ylabel('Number of Review')
plt.title('Price vs Number of Reviews')
plt.show()
```

After analysis we found that most people prefer to stay in place where price is cheap or less

**3. Which hosts are the busiest and why?**

```python
busiest_hosts = new_df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_hosts = busiest_hosts.sort_values(by='number_of_reviews',ascending=False).head(10)
busiest_hosts
```

```python
name= busiest_hosts['host_name']
reviews=busiest_hosts['number_of_reviews']

fig = plt.figure(figsize=(10,5))

#creat the bar plot
plt.bar(name, reviews, color='green',width=0.4)

plt.xlabel('Name of the Host')
plt.ylabel('Number of Review')
plt.title('Busiest Hosts')
plt.show()
```

**4. Is there any noticeable difference of traffic among different areas and what could be the reason for it?**


```python
traffic_areas = new_df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights',ascending=False)
traffic_areas
```

```python
room_type = traffic_areas['room_type']
stayed = traffic_areas['minimum_nights']

fig = plt.figure(figsize=(10,5))

#creat bar plot
plt.bar(room_type, stayed, color='green',width=0.4)

plt.xlabel('Room Type')
plt.ylabel('Minmum number of nights stayed')
plt.title('Traffic Areas')
plt.show()
```

This above Analysis stays that people are preferring entire home/apt or private room which are present in Manhattan,Brooklyn,Queens and prople are listing in less price

CONCLUSION:

1.In that particular neighbourhood only peoples who prefer to stay in entire home or apartment they are going to stay bit longer.

2.In compared to home and apartment peoples who prefer to stay in private room thet don't stay longer.

3.Most people prefer to pay less price.

4.If there are more number of reviews for particular neighbourhood group that means that place is a tourist plac.

5.Those peoples are traveller who don't stay more then one nights.

**1.What is the average prefered price by customers according to the location?**

```python
#we apply dropna() function to entire dataset and drop null values with dropna() function
airbnb_df=df.dropna()
```

```python
airbnb_df.info()
```

```python
airbnb_df.head()
```

```python
airbnb_df.describe()
```

```python
#sum of all null values in dataset
airbnb_df.isnull().sum()
```

```python
#average price of property according to the location
avg_preffered_price_df=airbnb_df.groupby(['neighbourhood_group','room_type'], as_index=False)['price'].mean().rename(columns={'neighbourhood_group':'Location','Price':'Average_Price'})
avg_preffered_price_df
```

```python
#Unstack the group by information for plot the graph
avg_preffered_price_df = airbnb_df.groupby(['neighbourhood_group','room_type'])['price'].mean().unstack()
avg_preffered_price_df
```

```python
# graph of average price of property according to location
avg_preffered_price_df.plot.bar()
```
